# 3.Create a 8x8 matrix and fill it with a checkerboard pattern

```python
import numpy as np
x = np.ones((4,4))
print("Checkerboard pattern:")
x = np.zeros((8,8),dtype=int)
x[1::2,::2] = 1
x[::2,1::2] = 1
print(x)
```

# 4. Normalize a 5x5 random matrix

```python
import numpy as np
x= np.random.random((5,5))
print("Original Array:")
print(x)
xmax, xmin = x.max(), x.min()
x = (x - xmin)/(xmax - xmin)
print("After normalization:")
print(x)
```

# 5.  How to find common values between two arrays?

```python
import numpy as np
array1 = np.array([0, 10, 20, 44, 60])
print("Array1: ",array1)
array2 = [20, 44, 55]
print("Array2: ",array2)
print("Common values between two arrays:")
print(np.intersect1d(array1, array2))
```

# 6.How to get the dates of yesterday, today and tomorrow?

```python
import numpy as np
yesterday = np.datetime64('today') - np.timedelta64(1)
today     = np.datetime64('today')
tomorrow  = np.datetime64('today') + np.timedelta64(1)
print(yesterday)
print(today)
print(tomorrow)
```

# 12.How to get the diagonal of a dot product?

```python
import numpy as np
A = np.random.uniform(0,1,(15,15))
B = np.random.uniform(0,1,(15,15))
print(A)
print(B)

prod1=np.diag(np.dot(A, B))


prod2=np.sum(A * B.T, axis=1)

print("diagonal of a dot product method 1 is....",prod1)
print("diagonal of a dot product method 2 is....",prod2)
```

# 15.How to create a record array from a regular array?

```python
import numpy as np
X = np.array([("Namasthe", 4.7, 2),
              ("India", 3.5, 2)])
Y = np.core.records.fromarrays(X.T,
                               names='col1, col2, col3',
                               formats = 'S8, f8, i8')
print(Y)
```
